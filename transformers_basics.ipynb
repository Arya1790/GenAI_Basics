## Sentiment analysis

 Creates pipelines for various natural language processing tasks, including sentiment analysis, text generation, named entity recognition, etc.

 By default, the model downloaded for this pipeline is called "distilbert-base-uncased-finetuned-sst-2-english"

 When typing this command for the first time, a pretrained model and its tokenizer are downloaded and cached. The tokenizer's job is to preprocess the text for the model, which is then responsible for making predictions. The pipeline groups all of that together, and post-process the predictions to make them readable.


```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
classifier("We are very happy to show you the 🤗 Transformers library.")
```

```python
classifier("The pizza is not great but the desert is awesome")
```

```python
results = classifier(['Today is a good day to start learning transformers.', "I hope you don't hate it"])
for result in results:
  print(f"label: {result['label']}, with score {round(result['score'],4)}" )
```

We can also replace that name by a local folder where you have saved a pretrained model and also pass a model object and its associated tokenizer.

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
# This model only exists in PyTorch, so we use the `from_pt` flag to import that model in TensorFlow.
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
classifier("Its a bright and sunny day")
```
